### 4. Conditional selection

#### Create a boolean mask for "bill_amt" less than 35000

````python
import numpy as np
import pandas as pd

mask = df_copy.loc[:, "bill_amt"] < 35000
print(mask)
````

#### Use boolean mask to get those rows which have "bill_amt" less than 3500

````python
df_copy[df_copy.loc[:, "bill_amt"] < 3500] 
````

#### Find rows with the column "name" equal to "Price"

````python
df_copy[df_copy.loc[:, "name"] == "Price"] 
````

#### Find rows where the "update_at" column is before 2018

````python
df_copy["updated_at"] = pd.to_datetime(df_copy["updated_at"])
df_copy[df_copy["updated_at"] < "2018"]
````

#### Select the rows where the "name" column is "Hackett" and the "bill_amt" is greater than 25000

````python
df_copy[(df_copy["name"] == "Hackett") & (df_copy["bill_amt"] > 25000)]
````
